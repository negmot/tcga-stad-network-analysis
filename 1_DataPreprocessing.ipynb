```python
import pandas as pd

# ✅ Local path to raw counts CSV file
file_path = r"C:\Users\Neg\RNA_seq_python_GC\raw_counts_backup.csv"

# ✅ Load raw counts
counts_df = pd.read_csv(file_path, index_col=0)

# ✅ Filter genes with count >1 in at least 5 samples
filtered_df = counts_df[counts_df.gt(1).sum(axis=1) >= 5]

# ✅ Save filtered data
filtered_path = r"C:\Users\Neg\RNA_seq_python_GC\filtered_counts.csv"
filtered_df.to_csv(filtered_path)

# ✅ Preview
print("Filtered count matrix shape:", filtered_df.shape)
filtered_df.head()
```

```python
import pandas as pd
import numpy as np

# Load filtered count matrix
filtered_path = r"C:\Users\Neg\RNA_seq_python_GC\filtered_counts.csv"
filtered_df = pd.read_csv(filtered_path, index_col=0)

# Apply log2 transformation with pseudocount
log2_df = np.log2(filtered_df + 1)

# Save the log2 normalized data
log2_path = r"C:\Users\Neg\RNA_seq_python_GC\log2_normalized_counts.csv"
log2_df.to_csv(log2_path)

# Preview
log2_df.head()
```
