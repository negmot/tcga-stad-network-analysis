```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the file
file_path = r"C:\Users\Neg\TCGA_GC_project\Network_analysis\6500\Cytohubba\CytoCluster\Top20_Reactome_pathways_for_heatmap.xlsx"
df = pd.read_excel(file_path, engine="openpyxl")

# Strip extra spaces from column names
df.columns = df.columns.str.strip()

# Fill down the 'Cluster Rank' column
df["Cluster Rank"] = df["Cluster Rank"].ffill().astype(int)

# Strip leading/trailing spaces from pathways
df["Pathways"] = df["Pathways"].str.strip()

# Create binary matrix (1 = pathway appears in that cluster)
heatmap_df = pd.crosstab(df["Pathways"], df["Cluster Rank"])

# Plot heatmap
plt.figure(figsize=(10, 20))
sns.heatmap(heatmap_df, cmap="Blues", cbar_kws={"label": "Pathway Presence"}, linewidths=0.5, linecolor='gray')

plt.title("Reactome Pathway Enrichment Across Top Clusters", fontsize=14)
plt.xlabel("Cluster Rank")
plt.ylabel("Reactome Pathway")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Define input and output paths
input_path = r"C:\Users\Neg\TCGA_GC_project\Network_analysis\6500\Cytohubba\CytoCluster\Top20_Reactome_pathways_for_heatmap.xlsx"
output_dir = r"C:\Users\Neg\TCGA_GC_project\Network_analysis\6500\Cytohubba\CytoCluster"
output_file = os.path.join(output_dir, "Reactome_Pathway_Cluster_Heatmap.png")

# Load the Excel file
df = pd.read_excel(input_path, engine="openpyxl")

# Clean column names and values
df.columns = df.columns.str.strip()
df["Cluster Rank"] = df["Cluster Rank"].ffill().astype(int)
df["Pathways"] = df["Pathways"].str.strip()

# Create binary matrix of pathway occurrences
heatmap_df = pd.crosstab(df["Pathways"], df["Cluster Rank"])

# Plot heatmap
plt.figure(figsize=(10, 20))
sns.heatmap(heatmap_df, cmap="Blues", cbar_kws={"label": "Pathway Presence"}, linewidths=0.5, linecolor='gray')

# Customize plot
plt.title("Reactome Pathway Enrichment Across Top Clusters", fontsize=14)
plt.xlabel("Cluster Rank")
plt.ylabel("Reactome Pathway")
plt.tight_layout()

# Save heatmap
plt.savefig(output_file, dpi=300)
plt.show()

print("✅ Heatmap saved to:", output_file)
```
