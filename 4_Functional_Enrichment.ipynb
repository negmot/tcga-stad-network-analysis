```python
import pandas as pd
import gseapy as gp
import matplotlib.pyplot as plt
import os
import numpy as np

# Define paths
input_path = r"C:\Users\Neg\TCGA_GC_project\Top200_HighConfidence_DEGs_with_Symbols.xlsx"
supplementary_folder = r"C:\Users\Neg\TCGA_GC_project\Supplementary_files_4_manuscript_submission"
figures_folder = r"C:\Users\Neg\TCGA_GC_project\Figures_4_manuscript_Submission"

# Make sure output folders exist
os.makedirs(supplementary_folder, exist_ok=True)
os.makedirs(figures_folder, exist_ok=True)

# Load DEGs
df = pd.read_excel(input_path)

# Separate upregulated and downregulated genes
up_genes = df[df['log2FC'] > 0]['Gene_Symbol'].dropna().tolist()
down_genes = df[df['log2FC'] < 0]['Gene_Symbol'].dropna().tolist()

# Define a function to run enrichment for multiple libraries
def run_enrichment_combined(gene_list, label):
    gene_sets = {
        "KEGG_2021_Human": "KEGG",
        "GO_Biological_Process_2021": "GO_BP",
        "GO_Molecular_Function_2021": "GO_MF",
        "GO_Cellular_Component_2021": "GO_CC"
    }
    
    all_results = []

    for lib, short_name in gene_sets.items():
        enr = gp.enrichr(
            gene_list=gene_list,
            gene_sets=lib,
            organism='Human',
            cutoff=0.05,
            no_plot=True,
            outdir=None,
        )
        if enr.res2d is not None and not enr.res2d.empty:
            temp = enr.res2d.copy()
            temp['Source'] = short_name
            all_results.append(temp)

    if not all_results:
        print(f"No enrichment terms found for {label}.")
        return

    combined_df = pd.concat(all_results, ignore_index=True)
    
    # Save combined enrichment table
    combined_df.to_excel(os.path.join(supplementary_folder, f"{label}_Combined_Enrichment.xlsx"), index=False)

    # Select top 10 terms per source
    top_terms = combined_df.groupby('Source').apply(lambda x: x.nsmallest(10, 'Adjusted P-value')).reset_index(drop=True)

    # Plotting
    plt.figure(figsize=(10, 8))
    colors = {
        'KEGG': 'red',
        'GO_BP': 'blue',
        'GO_MF': 'green',
        'GO_CC': 'purple'
    }
    term_labels = top_terms['Term'].apply(lambda x: x.split(' (')[0])  # Clean GO IDs from labels
    
    plt.barh(term_labels, -np.log10(top_terms['Adjusted P-value']), color=top_terms['Source'].map(colors), edgecolor='black')
    plt.xlabel('-log10(FDR)', fontsize=12)
    plt.title(f'Functional Enrichment of {label} DEGs', fontsize=14, weight='bold')
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.gca().invert_yaxis()
    
    # Legend
    import matplotlib.patches as mpatches
    legend_handles = [mpatches.Patch(color=col, label=source) for source, col in colors.items()]
    plt.legend(handles=legend_handles, title="Category", loc='best')
    
    # Save plot
    plt.tight_layout()
    plt.savefig(os.path.join(figures_folder, f"{label}_Combined_Enrichment.png"), dpi=300)
    plt.close()
    
    print(f"✅ {label} Combined Enrichment Done!")

# Run for Upregulated and Downregulated separately
run_enrichment_combined(up_genes, "Upregulated")
run_enrichment_combined(down_genes, "Downregulated")
```
