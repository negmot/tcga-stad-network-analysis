```python
import pandas as pd

# 🔍 Load the clinical metadata
clinical_path = r"C:\Users\Neg\RNA_seq_python_GC\tcga_stad_clinical.tsv.gz"
clinical_df = pd.read_csv(clinical_path, sep='\t', compression='gzip')

# 🧾 Preview the structure
print("Clinical data shape:", clinical_df.shape)
clinical_df.head()
```

```python
# Select key clinical variables
columns_of_interest = [
    'sample',
    'gender.demographic',
    'race.demographic',
    'ethnicity.demographic',
    'days_to_birth.diagnoses',
    'tumor_stage.diagnoses',
    'vital_status.diagnoses',
    'days_to_death.diagnoses',
    'days_to_last_follow_up.diagnoses'
]

# Subset and rename for simplicity
clinical_cleaned = clinical_df[columns_of_interest].copy()
clinical_cleaned.columns = [
    'sample_id', 'gender', 'race', 'ethnicity',
    'days_to_birth', 'tumor_stage',
    'vital_status', 'days_to_death', 'days_to_last_followup'
]

# Save cleaned clinical data
cleaned_path = r"C:\Users\Neg\RNA_seq_python_GC\clinical_cleaned.csv"
clinical_cleaned.to_csv(cleaned_path, index=False)

# Confirm shape
print("✅ Cleaned clinical data shape:", clinical_cleaned.shape)
clinical_cleaned.head()
```

```python
import pandas as pd

# Load clinical + filtered expression data
clinical = pd.read_csv(r"C:\Users\Neg\RNA_seq_python_GC\clinical_cleaned.csv")
expr = pd.read_csv(r"C:\Users\Neg\RNA_seq_python_GC\filtered_counts.csv", index_col=0)

# Standardize sample IDs (first 15 characters)
expr.columns = expr.columns.str.slice(0, 15)

# Find shared sample IDs
common_ids = list(set(clinical['sample_id']) & set(expr.columns))

# Subset both datasets
expr_matched = expr[common_ids]
clinical_matched = clinical[clinical['sample_id'].isin(common_ids)].copy()

# Save to local folder
expr_matched.to_csv(r"C:\Users\Neg\RNA_seq_python_GC\expression_matched.csv")
clinical_matched.to_csv(r"C:\Users\Neg\RNA_seq_python_GC\clinical_matched.csv", index=False)

print("✅ Matched expression and clinical files saved.")
```
