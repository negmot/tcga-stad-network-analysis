```python
import pandas as pd
from mygene import MyGeneInfo

# Load your CSV file
df = pd.read_csv(r"C:\Users\Neg\TCGA_GC_project\DEGs_MWU_filtered_log2fc.csv")

# Remove Ensembl version suffix (e.g., ".14")
df["Gene_clean"] = df["Gene"].astype(str).str.split(".").str[0]

# Query gene symbols from Ensembl IDs
mg = MyGeneInfo()
ensembl_ids = df["Gene_clean"].tolist()
res = mg.querymany(ensembl_ids, scopes="ensembl.gene", fields="symbol", species="human")
res_df = pd.DataFrame(res)

# Remove entries with no match
res_df = res_df[res_df["notfound"] != True]

# Merge original data with symbols
merged = df.merge(res_df[["query", "symbol"]], left_on="Gene_clean", right_on="query")

# Save selected columns to Excel
output_path = r"C:\Users\Neg\TCGA_GC_project\All_DEGs_GeneSymbols_for_STRING.xlsx"
merged[["symbol", "Gene", "log2FC_x", "p_value", "FDR", "abs_log2FC"]].drop_duplicates().to_excel(output_path, index=False)

print(f"✅ Excel file saved to: {output_path}")
```
